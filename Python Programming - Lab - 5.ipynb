### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s=input("Enter string")
ucount=0
lcount=0
dcount=0
scount=0
for i in s:
    if i.isupper():
        ucount+=1
      
    elif i.islower():
        lcount+=1

    elif i.isdigit():
        dcount+=1

    else:
        scount+=1

print("uppercase=",ucount,"lowercase=",lcount,"digit=",dcount,"special=",scount)

        
    
```

### 8) Replace Vowels
- `Programming` → `Pr*gr*mm*ng`
- `Education` → `*d*c*t**n`
- `Sky` → `Sky`

```python
s=input("Enter string")
for i in s:
    if i=='a':
       s=s.replace('a','*')
    if i=='e':
        s=s.replace('e','*')
    if i=='i':
       s=s.replace('i','*')
    if i=='o':
        s=s.replace('o','*')
    if i=='u':
        s=s.replace('u','*')
print(s)        
        
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s=input("Enter string")
ans=""
count=1
for i in range(1,len(s)):
    if s[i]==s[i-1]:
        count+=1
    else:
        # ans=count.join(i)
        ans=ans+s[i-1]+str(count)
        count=1
ans=ans+s[-1]+str(count)       
print(ans)        
```

### 10) Toggle Case (without using str.swapcase())

### 15) Custom Split
- `apple,banana,grape`
- `10|20|30`
- `A-B-C-D`

```python
text = input("Enter string: ")
sep = input("Enter separator: ")

word = ""
result = []

for ch in text:
    if ch == sep:
        result.append(word)
        word = ""
    else:
        word += ch

result.append(word)  

print(result)
```
